```python
import os

def create_folder(folder_name):
    if not os.path.exists(folder_name):
        os.makedirs(folder_name)
        print(f"Folder '{folder_name}' created successfully.")
    else:
        print(f"Folder '{folder_name}' already exists.")

# Example usage
folder_name1 = 'Dogs/source'
folder_name2 = 'Dogs/target'
create_folder(folder_name1)
create_folder(folder_name2)
```

```python
import os
import shutil
import glob

def copy_images(source_dir, target_dir):
    # Create the target directory if it doesn't exist
    if not os.path.exists(target_dir):
        os.makedirs(target_dir)

    # Traverse all subdirectories and copy image files
    for root, dirs, files in os.walk(source_dir):
        for file in files:
            # Check if the file is an image (modify the extensions list as needed)
            if file.lower().endswith(('.jpg', '.jpeg', '.png', '.gif')):
                source_path = os.path.join(root, file)
                target_path = os.path.join(target_dir, file)
                shutil.copy2(source_path, target_path)
                print(f"Copied '{file}' to '{target_dir}'")
```

```python
sourcef = '/content/Images'  # Path to the source folder
destinationf = '/content/Dogs/target'  # Path to the destination folder
copy_images(sourcef, destinationf)
```

```python
import os
from PIL import Image

def open_images(folder_path):
    # Iterate over all files in the folder
    for filename in os.listdir(folder_path):
        file_path = os.path.join(folder_path, filename)

        # Check if the file is an image
        if os.path.isfile(file_path) and filename.lower().endswith(('.jpg', '.jpeg', '.png', '.gif')):
            try:
                # Open the image using Pillow
                image = Image.open(file_path)

                # Close the image file
                image.close()

                print(f"Opened image: {filename}")
            except (IOError, SyntaxError) as e:
                # Delete the file if there's an error opening it
                os.remove(file_path)
                print(f"Deleted image: {filename} ({e})")

# Provide the folder path containing the images
folder_path = "/content/Dogs/target"

# Open and process the images, deleting any that raise an error
open_images(folder_path)
```

```python
def resize_images(folder_path, output_path, size):
    # Create the output directory if it doesn't exist
    os.makedirs(output_path, exist_ok=True)

    # Iterate over all files in the folder
    for filename in os.listdir(folder_path):
        file_path = os.path.join(folder_path, filename)

        # Check if the file is an image
        if os.path.isfile(file_path) and filename.lower().endswith(('.jpg', '.jpeg', '.png', '.gif')):
            # Open the image using Pillow
            image = Image.open(file_path)

            # Resize the image
            resized_image = image.resize((size, size))

            resized_image = resized_image.convert('RGB')

            # Construct the output file path
            output_file_path = os.path.join(output_path, filename)

            # Save the resized image
            resized_image.save(output_file_path)

            # Close the image file
            image.close()

            print(f"Resized image: {filename}")

# Provide the folder path containing the images
folder_path = "/content/Dogs/target"

# Provide the output folder path where the resized images will be saved
output_path = "/content/Dogs/target"

# Provide the desired size for resizing (512x512 in this case)
size = 512

# Resize the images
resize_images(folder_path, output_path, size)
```

```python
import os
import cv2

def apply_canny_edge_detector(source_folder, target_folder, threshold1=100, threshold2=200):
    # Create the target folder if it doesn't exist
    if not os.path.exists(target_folder):
        os.makedirs(target_folder)

    # Iterate over all files in the source folder
    for filename in os.listdir(source_folder):
        source_path = os.path.join(source_folder, filename)
        target_path = os.path.join(target_folder, filename)

        # Check if the file is an image (modify the extensions list as needed)
        if os.path.isfile(source_path) and filename.lower().endswith(('.jpg', '.jpeg', '.png', '.gif')):
            # Read the image
            image = cv2.imread(source_path, cv2.IMREAD_GRAYSCALE)

            # Apply Canny edge detection
            edges = cv2.Canny(image, threshold1, threshold2)

            # Save the result in the target folder
            cv2.imwrite(target_path, edges)

            print(f"Processed '{filename}' and saved in '{target_folder}'")

# Example usage
source_folder = "/content/Dogs/target"
target_folder = "/content/Dogs/source"

apply_canny_edge_detector(source_folder, target_folder)
```

```python
import os
import json
import requests
from PIL import Image
from transformers import BlipProcessor, BlipForConditionalGeneration

# Initialize the Blip image captioning model
# List to store the image locations and captions
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base", torch_dtype=torch.float16).to("cuda")

def generate_captions(source_folder, json_file, model, processor):
    image_captions = []

    # Iterate over all files in the source folder
    for filename in os.listdir(source_folder):
        image_path = os.path.join(source_folder, filename)

        # Read the image
        raw_image = Image.open(image_path).convert('RGB')

        # Generate captions for the image
        text = "a photography of"
        inputs = processor(raw_image, text, return_tensors="pt").to("cuda", torch.float16)
        out = model.generate(**inputs)
        caption = processor.decode(out[0], skip_special_tokens=True)

        # Append image location and caption to the list
        image_caption = {
            "source": filename,
            "target": filename,
            "prompt": caption
        }
        image_captions.append(image_caption)
        # print('Done!')

    # Save the image locations and captions in a JSON file
    with open(json_file, 'w') as f:
        json.dump(image_captions, f, indent=4)

    print(f"Image captions saved in '{json_file}'")

# Example usage
source_folder = "/content/Dogs/target"
json_file = "/content/Dogs/prompt.json"

generate_captions(source_folder, json_file, model, processor)
```

```python
input_file = "prompt - Copy.json"
output_file = "prompt-new.json"

# Read the input file
with open(input_file, "r") as file:
    data = json.load(file)

# Write each dictionary in a new line in the output file
with open(output_file, "w") as file:
    for item in data:
        # Convert the dictionary to a string
        item_str = json.dumps(item)
        # Write the string to the file with a newline character
        file.write(item_str + "\n")
```
